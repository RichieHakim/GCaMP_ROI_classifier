```python
import glob
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
files = glob.glob('./*.csv')
```

```python
df_lst = {}
for file in files:
    df_lst[file] = pd.read_csv(file)
```

```python
val = {'SYT' in _:df_lst[_].set_index('Unnamed: 0') for _ in df_lst if 'val' in _ and 'efficient2-only' in _}
tr = {'SYT' in _:df_lst[_].set_index('Unnamed: 0') for _ in df_lst if 'train' in _ and 'efficient2-only' in _}
```

```python
syt_tr = tr[True]
rch_tr = tr[False]
syt_val = val[True]
rch_val = val[False]
```

```python
plt.plot(syt_tr.max(axis=1), label='Train')
plt.plot(syt_val.max(axis=1), label='Val')
plt.title('SYT Accuracy vs. Epoch')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.grid()
plt.legend()
```

```python
plt.plot(syt_tr.max(axis=1), label='Train')
plt.plot(rch_val.max(axis=1), label='Val')
plt.title('Rich Accuracy vs. Epoch')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.grid()
plt.legend()
```
