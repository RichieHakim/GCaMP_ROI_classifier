```python
import torch
import torchvision
```

```python
def get_num_params(mod):
    sm = 0
    for param in mod.parameters():
        mult = 1
        for dim in param.shape:
            mult *= dim
        sm += mult
    return sm
```

```python
tmp = torchvision.models.efficientnet_b0(pretrained=False)
get_num_params(tmp)
```

```python
tmp = torchvision.models.resnet18(pretrained=False)
get_num_params(tmp)
```
